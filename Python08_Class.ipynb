```python
class Car:
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year
    def display_info(self):
        print(f"{self.year} {self.brand} {self.model}")
car1 = Car("Toyota", "Corolla", 2020)
car1.display_info()
```

```python
class Person:
    def __init__(self, name):
        self.name = name
    def greet(self):
        return "Hello, " + self.name
    def welcome(self):
        message = self.greet()
        print(message + "! Welcome to our website.")
p1 = Person("Tobias")
p1.welcome()
```

```python
class Person:
  species = "Human" # Class property
    
  def __init__(self, name):
    self.name = name # Instance property

p1 = Person("Emil")
p2 = Person("Tobias")
print(p1.name)
print(p2.name)
print(p1.species)
print(p2.species)
```

```python
class Calculator:
  def add(self, a, b):
    return a + b
  def multiply(self, a, b):
    return a * b
calc = Calculator()
print(calc.add(5, 3))
print(calc.multiply(4, 7))
```
